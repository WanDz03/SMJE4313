Now that we have the path to the image, we can load it off the disk using the cv2.imread function on Line 1. The cv2.imread function returns a NumPy array representing the image.

Then, convert the image to grayscale image

Finally, displaying the actual image on our screen. The parameter is a reference to the image we convert before.

```
# Task 2
# Load the image and show it here :
image = cv2.imread(args["image"])
cv2_imshow(image)

# Task 3
# Convert image to grayscale

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Task 4
# Show the image
cv2_imshow(image_gray)
```

Line 2 computes the actual histogram. Go ahead and match the arguments of the code up with the function documentation above. We can see that our rst parameter is the grayscale image. A grayscale image has only one channel, hence we have a value of [0] for channels. We dont have a mask, so we set the mask value to None. We will use 256 bins in our histogram, and the possible values range from 0 to 256.

Finally, a call plt.plot() to plots our grayscale histogram.

```
# Construct a grayscale histogram
hist = cv2.calcHist([image_gray], [0], None, [256], [0, 256])

# Plot the histogram
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(hist)
plt.xlim([0, 256])

# Task 5
# Show the plot here :
plt.show()
```
