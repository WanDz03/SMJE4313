**Convert to CMYK**

```python
!pip install matplotlib
!pip install pillow

from PIL import Image
import matplotlib.pyplot as plt

#Step 2: Load an image using PIL
#image_url = 'https://example.com/your-image.jpg' # specify the URL to your image here
image_path = '/content/drive/MyDrive/Colab Notebooks/SMJE4313 Image Processing/canyon.jpg' # or specify the local path to your image here

# If using an image URL
# !wget {image_url} -O your-image.jpg
# rgb_image = Image.open('your-image.jpg')

# If using an image path
rgb_image = Image.open(image_path)

# Step 3: Convert the image from RGB to CMYK
cmyk_image = rgb_image.convert('CMYK')

# Step 4: Display the image
# Note that matplotlib will convert it back to RGB for display, but the data is in CMYK
plt.imshow(cmyk_image)
plt.axis('off')
plt.show()
```
